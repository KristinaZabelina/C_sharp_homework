```python
import pandas as pd
import seaborn as seab
```

```python
penguins = seab.load_dataset("penguins")
penguins.head()
```

```python
seab.scatterplot(data=penguins, x = 'bill_length_mm', y = 'flipper_length_mm', hue = 'sex', size = 'species', style = 'island')
```

```python
cols = ['species', 'island', 'bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g', 'sex']
penguins_2 = penguins[penguins['species'] == 'Adelie']

g = seab.PairGrid(penguins_2[cols])
g.map(seab.scatterplot)
```

```python
seab.heatmap(data = penguins.head(25)[['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm',	'body_mass_g']])
```
